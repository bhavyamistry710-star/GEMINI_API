```python
import os

#  Create an API client 

import google.generativeai as genai

GEMINI_API_KEY = os.environ.get('GEMINI_API_KEY')

genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel('gemini-2.5-flash')
config = genai.GenerationConfig(
     max_output_tokens=2048
)
```

```python


# Helper Function 1: Add a user message 
# Helper Function 1: Add a user message 
def add_user_message(history, content_text):
    history.append({
        'role': 'user',
        'parts': [content_text]  
    })

# Helper Function 2: Add an AI message
def add_assistant_message(history, response_text):
    history.append({
        'role': 'model',        
        'parts': [response_text]
    })

# Helper Function 3: Clean, dynamic dictionary execution wrapper
def chat(history, stop_sequences=None):
    generation_options = {
        "max_output_tokens": 2048,
        "stop_sequences": stop_sequences
    }
    
    response = model.generate_content(
        history, 
        generation_config=generation_options,
        stream=True
    )
    
    full_response = ""
    print("Assistant: ", end="", flush=True)
    for chunk in response:
        full_response += chunk.text
        print(chunk.text, end="", flush=True)
    print() # Adds a clean closing line breaks after streaming finishes
    return full_response


# --- RUNNING THE DATA TRACKING STREAM SESSION ---

history = []

# Turn 1
add_user_message(history, 'What is Gemini? Answer in three sentences.')
answer = chat(history)
add_assistant_message(history, answer)
print("-" * 50)

# Turn 2 (Perfect follow-up memory)
add_user_message(history, 'What is Gemini-3.5 Flash? Answer in one sentence.')
answer = chat(history)
add_assistant_message(history, answer)
print("-" * 50)

# View the beautifully structured history output list
print("\n📋 Full Conversation History State:")
import pprint
pprint.pprint(history)
```
